```python
def repeater(s1, s2, n):
  # this function repeats strings
  returner = s1 + s2
  returner = "_" + (returner * n) + "_"
  return returner
```

```python
repeater('AAA', 'x', 3)
```

```python
repeater('+', '--', 30)
```

```python
repeater('T', 'l_l', 100)
```

```python
import math

def roots(a, b, c):
  # this program finds roots
  delta = b**2 - (4 * a * c)
  ret = []
  if delta > 0:
    root1 = (-b + math.sqrt(delta)) / (2 * a)
    root2 = (-b - math.sqrt(delta)) / (2 * a)
    ret.append(root1)
    ret.append(root2)
  elif delta == 0:
    root = (-b) / (2 * a)
    ret.append(root)
  print(
    "The roots of the function f(x) = ({0}x^2) + ({1}x) + ({2}) are {3}.".format(
    a, b, c, ret    
    )
  )
```

```python
roots(1, 0, 0)
```

```python
roots(1, 0, 2)
```

```python
roots(1, 0, -1)
```

```python
roots(-3, 6, 9)
```

```python
def real_roots(a, b, c):
  # this program finds real roots
  delta = b**2 - (4 * a * c)
  return delta >= 0
```

```python
real_roots(1, 0, 0)
```

```python
real_roots(1, 0, 2)
```

```python
real_roots(1, 0, -1)
```

```python
real_roots(-3, 6, 9)
```

```python
def reverse(x):
  # This function reverses a 2 digit number.
  ret = int(x / 10)
  ret = ((x - (ret * 10)) * 10) + ret
  return ret
```

```python
reverse(26)
```
