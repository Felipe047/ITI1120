Repeat the exercise in the task 3 (version 2), where in addition you need to ask the user for their citizenship and if they are currently in prison convicted for a criminal offence. Your program should print a nice message telling the user if they are eligible to vote (i.e. if they are 18+, Canadian and do not live in prison convicted for a criminal offence, then they can vote. Otherwise not). You should modify function is_eligible so it takes to additional paramters as input. In particular the head of the function should be:  is_eligible(age, citizenship, prison)

You should define your function in a cell and your I/O(input and output, user interactions) in another cell.

```
# define your function here
def is_eligible(age, citizenship, prison):
     '''(int, str, str)->bool
     Returns true if a person is eligible to vote, and false otherwise
     '''
     ageb = age >= 18
     citb = citizenship.lower() == "y"
     prib = prison.lower() == "y"

     return ageb and citb and prib
 
```

```
# implement your I/O here
name=input("What is your name? ")
age= int(input("How old are you? "))
citizenship = input("Are you a canadian citizen?(Input y if yes) ")
prison = input("Are you in prison?(Input y if yes) ")


if is_eligible(age, citizenship, prison):
     print(name, ", you are eligible to vote")
else:
     print(name, ", you are ineligible to vote") 
```

```
# Your code goes here
def half_pyramid(num, char):
    ''' This program prints a half pyramid with a selected character'''
    num = int(num)
    char = str(char)
    for c in range(1, num + 1):
      print(char*c)
```

```
def pyramid(num, char):
    ''' This program prints a pyramid with a selected character'''
    num = int(num)
    char = str(char)

    for c in range(1, num + 1):
      printable = " " * (num - c) + char * (2*(c - 1) + 1) + " " * (num - c)
      print(printable)
```

```
half_pyramid(3, "&")
```

```
pyramid(10, "#")
```

```
number = int(input("Input a positive integer: "))
character = input("Input a character: ")

print("Your half pyramid looks like this:\n")

half_pyramid(number, character)

print("\nYour pyramid looks like this:\n")

pyramid(number, character)
```

# Exercise 4

```
def divisors1(n):
    '''(int)->list
    prints all the divisors of n
    Precondition: n is a positive integer'''
     
    div = []

    for c in range(1, n + 1):
      if n % c == 0:
        div.append(c)

    return div
```

```
def divisors(n):
    '''(int)->none
    prints all the divisors of n
    Precondition: n is a positive integer'''
     
    div = divisors1(n)

    print_v = ""
    for c in range(len(div)):
        if c == 0:
            print_v = print_v + str(div[c])
        else:
            print_v = print_v + ", " + str(div[c])

    print(print_v)
```

```
def prime(n):
     '''(int)->bool
     returns True if n is a prime, and False otherwise
     Precondition: n is a positive integer'''

     div = divisors1(n)

     return len(div) <= 2
     
```

```
def prime_v2(n):
    '''(int)->bool
    returns True if n is a prime, and False otherwise
    Precondition: n is a positive integer'''

    print_val = "1"

    for c in range(2, n + 1):
        if prime(c):
            print_val = print_val + ", " + str(c)
     
    print(print_val)
```

```
# Write your I/O here
# prompt for a number
number = int(input("Please input a positive integer: "))
# divisors:
print("The divisors of the number are:")
divisors(number)
# primality
if prime(number):
    print("{} is a prime number.".format(number))
else:
    print("{} is not a prime number.".format(number))

# all primies less than
print("All the primes up to this point are:")
prime_v2(number)
```
