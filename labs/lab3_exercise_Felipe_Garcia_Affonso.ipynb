```
### complete your code here ###
def pay(hourly_pay, hours_worked):
  '''pay(hourly_pay, hours_worked) -> payday'''
  hourly_pay = float(hourly_pay)
  hours_worked = float(hours_worked)

  if hours_worked < 45:
    return hours_worked * hourly_pay
  elif hours_worked <= 70:
    return hourly_pay * (45 + (hours_worked - 45) * 1.5)
  else:
    return hourly_pay * (45 + 25 * 1.5+ (hours_worked - 70) * 2)
### complete your code here ###
```

```
### complete your code here ###
def rps(play2, play1):
  '''Rock paper scissors. returns 1 0 or -1'''
  if play1 == play2:
    return 0
  elif play1 == "R" and play2 == "S":
    return 1
  elif play1 == "S" and play2 == "P":
    return 1
  elif play1 == "P" and play2 == "R":
    return 1
  else:
    return -1
### complete your code here ###
```
