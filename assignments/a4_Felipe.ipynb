Question 1:
Write a function called tally_scores that takes as a parameter the name of a file that contains a series of student records and that prints a summary of each student record.  A student record will begin with a name followed by a sequence of integer test scores.  The name is guaranteed to be one word and at the beginning of the line.  You may assume that each student has at least one test score.  Your function should produce two lines of output for each student: one showing the student's name and test scores and a second line showing the average score.
For example, if a file called records.txt contains the following:

	John 71 83 94 
	Sally 94 85 
	Fred 90 95 82 85
  and the following call is made:
  
    tally_scores(records)
 the following output should be produced:

	John: 71 83 94
	average = 82.66666666666667
	Sally: 94 85
	average = 89.5
	Fred: 90 95 82 85
	average = 88.0

You are to exactly reproduce the format of this output.  You may not construct any extra data structures to solve this problem.


```python
##########################
# Question 1 Solution
##########################
from google.colab import files

def tally_scores(file_name):
  '''
  (File Path) -> None
  Prints the data from the file formated.
  '''
  data = open(file_name).read().split("\n")

  for i in data:
    separated_data = i.split(" ")

    print(separated_data[0] + ": ", end="")

    avg = 0
    num_of_num = 0
    for n in range(1, len(separated_data)):
      try:
        number = int(separated_data[n])
        avg += number
        num_of_num += 1
        
        print("{0} ".format(number), end="")
      except:
        None
    avg = avg / num_of_num
    print("\naverage = {0}".format(avg))
```

Question 2:
Write a function called underline that takes a file name as a parameter and that prints the words one per line with certain words underlined.  The words to be underlined all begin with a period.  The period should not be printed.  You should print the text that follows the period on a line by itself followed by a line of alternating dashes and commas equal in length to the text that follows the period. 

Q02.PNG


Notice that input lines can be blank lines.  You may not construct any extra data structures to solve this problem other than strings.


```python
##########################
# Question 2 Solution
##########################
def underline(file_name):
  '''
  (File Path) -> None
  underlines words with . before
  '''
  data = open(file_name).read().split(" ")

  for i in data:
    for n in i.split("\n"):
      if len(n) == 0 or n[0] != ".":
        print(n)
      else:
        print(n[1:])
        for e in range(1, len(n)):
          if e % 2 == 0:
            print(",", end="")
          else:
            print("-", end="")
        print()
  
```

Question 3:
Write a function called merge_dict that takes two dictionaries, merge them, and returns a new dictionary. The returned dictionary should contain all the keys of the passed dictionaries. Moreover, the values of the common keys will be merged together in one list. However, the values of the unique keys should stay in the same format they have appeared in the input dictionaries.
For example if you had the following dictionaries:


d1 = {1:"Moh", 2:"Aseel", 3:"Batoul", 4:["Taylor", "Chan", "Laura"]}

d2 = {1: "Nahid", 2:"Carter", 3:"John", 6:"Odak"} 

Output:

{1: ['Nahid', 'Moh'], 2: ['Carter', 'Aseel'], 3: ['John', 'Batoul'], 6: 'Odak', 4: ['Taylor', 'Chan', 'Laura']}

Another Example:

d1 = {1:"Moh", 2:"Aseel", 3:"Batoul", 4:["Taylor", "Chan", "Laura"]}

d2 = {1: "Nahid", 2:"Carter", 5:"John", 4:"Odak"}


output:

{1: ['Nahid', 'Moh'], 2: ['Carter', 'Aseel'], 5: 'John', 4: ['Taylor', 'Chan', 'Laura', 'Odak'], 3: 'Batoul'}

```python

##########################
# Question 3 Solution
##########################
def merge_dictionaries(a,b):
  '''
  (dict, dict) -> dict
  '''
  a_keys = a.keys()
  b_keys = b.keys()

  dic = {}

  for i in a_keys:
    try:
      a_val = a[i]
      b_val = b[i]

      if isinstance(a_val, list) and isinstance(b_val, list):
        for n in a_val:
          b_val.append(n)
        dic[i] = b_val
      elif isinstance(b_val, list):
        b_val[0].append(a_val)
        dic[i] = b_val
      elif isinstance(a_val, list):
        a_val.append(b_val)
        dic[i] = a_val
      else:
        dic[i] = [a_val, b_val]
    except:
      dic[i] = a[i]

  for i in b_keys:
    try:
      a_val = a[i]
      b_val = b[i]
      None
    except:
      dic[i] = b[i]

  return dic
```

```python
d1 = {1:"Moh", 2:"Aseel", 3:"Batoul", 4:["Taylor", "Chan", "Laura"]}
d2 = {1: "Nahid", 2:"Carter", 3:"John", 6:"Odak"} 

### Question 3 Solution ~ test command 1 ###
result = merge_dictionaries(d1, d2)
print(result)
```

Consider a simple school management system that consists of two classes. A Student class and a Section class. 
The Student class has the following attributes and behaviors:

Implement the two classes. Test your implementation by instantiation three objects of the Student class and one object of the Section class. For example, if you write the code below in the main section of the program and execute your code, you should get the below output:

```python
##########################
# Question 4 Solution
##########################
class Student():
  def __init__(self, name, id, score):
    '''
    (str, str, int) -> None
    '''
    self.name = name
    self.id = id
    self.score = score
  
  def getStdName(self):
    return self.name

  def getStdId(self):
    return self.id

  def getStdScore(self):
    return self.score

  def __str__(self):
    print(self.name + " " + self.id + " " + str(self.score))

class Section(Student):

  def __init__(self, name, section):
    '''
    (str, str) -> None
    '''
    self.name = name
    self.section = section
    self.stu_list = []

  def add(self, student):
    '''
    (Student) -> None
    '''
    self.stu_list.append(student)

  def remove(self, student):
    '''
    (Student) -> None
    '''
    try:
      self.stu_list.remove(student)
    except:
      None

  def updateScore(self, student, score):
    '''
    (Student, int) -> None
    '''
    for i in self.stu_list:
      if i == student:
        student.score = score

  def displaySection(self):
    print("List of all students in section:")
    for i in self.stu_list:
      i.__str__()
    
  def averageScore(self):
    avg = 0
    num_of_num = 0
    for i in self.stu_list:
      avg += i.score
      num_of_num += 1
    
    avg = avg / num_of_num

    print("Average of entire section: {0:.2f}".format(avg))
```
