Problem 1

```python
def same_pattern(str1, str2):
  '''
  same_pattern(string, string) -> boolean
  This function analizes the pattens of 2 strings to see if they are the same.
  '''
  if len(str1) != len(str2):
    return False

  for n in range(len(str1)):
    char1 = str1[n]
    char2 = str2[n]
    for n1 in range(n, len(str1)):
      if (str1[n1] == char1) != (str2[n1] == char2):
        return False
  
  return True
```

```python
print(same_pattern("", ""))
print(same_pattern("a", "x"))
print(same_pattern("a", "ab"))
print(same_pattern("ab", "ab"))
print(same_pattern("aa", "xy"))
print(same_pattern("aba", "+-+"))
print(same_pattern("---", "aba"))
print(same_pattern("abcabc", "zodzod"))
print(same_pattern("abcabd", "zodzoe"))
print(same_pattern("abcabc", "xxxxxx"))
print(same_pattern("aaassscccn", "aaabbbcccd"))
print(same_pattern("asasasasas", "xyxyxyxyxy"))
print(same_pattern("ascneencsa", "aeiouuoiea"))
print(same_pattern("aaassscccn", "aaabbbcccd"))
print(same_pattern("asasasasas", "xxxxxyyyyy"))
print(same_pattern("ascneencsa", "aeiouaeiou"))
print(same_pattern("aaassscccn", "xxxyyyzzzz"))
print(same_pattern("aaasssiiii", "gggdddfffh"))
```

Problem 2

```python
def is_unique(list_of_int):
  '''
  same_pattern(list) -> boolean
  This function returns false if values inside repeat and true otherwise.
  '''

  for n in range(len(list_of_int)):
    value = list_of_int[n]
    for n1 in range(n + 1, len(list_of_int)):
      if list_of_int[n1] == value:
        return False
  
  return True
```

```python
print(is_unique([3, 8, 12, 2, 9, 17, 43, -8, 46, 203, 14, 97, 10, 4]))
print(is_unique([4, 7, 2, 3, 9, 12, -47, -19, 308, 3, 74]))
print(is_unique([1, 2, 4]))
print(is_unique([1, 1, 1]))
```

```python
def longest_sorted_sequence(list_of_int):
  '''
  same_pattern(list) -> boolean
  This function the biggest increasing sequence of numbers inside of it.
  '''

  if len(list_of_int) <= 1:
    return len(list_of_int)

  return_val = 0
  current_sort = 0

  for n in range(len(list_of_int) - 1):
    if list_of_int[n] <= list_of_int[n + 1]:
      current_sort += 1
    else:
      current_sort = 0
    
    if return_val < current_sort:
      return_val = current_sort
  
  return return_val + 1
```

```python
print(longest_sorted_sequence([]))
print(longest_sorted_sequence([1204]))
print(longest_sorted_sequence([3, 8, 10, 1, 9, 14, -3, 0, 14, 207, 56, 98, 12]))
print(longest_sorted_sequence([17, 42, 3, 5, 5, 5, 8, 2, 4, 6, 1, 19]))
```
