```python
def series_sum():
  '''
  This functio asks the user for a non negative integer and outputs
  1000 + 1 / (1 ** 2) + 1 / (2 ** 2) + 1 / (3 ** 2) ...
  + 1 / (n ** 2)
  '''
  number = int(input("Please enter a non-negative intger: "))

  if number < 0:
    return None

  ret_val = 1000

  for n in range(number):
    ret_val += 1 / ((n + 1) ** 2)
  
  return ret_val
```

```python
def encrypt(s):
  '''This function gets an string and encrypts it using a secret algorithm.'''
  
  string = str(s)

  first_half = string[:int(len(string) / 2)]
  second_half = string[int(len(string) / 2): len(string)]
  reverse_half = second_half[::-1]

  ret = ""

  for n in range(len(reverse_half)):
    ret += reverse_half[n]
    
    if n < len(first_half):
      ret += first_half[n]
    
  return ret
```

```python
def month_apart(m1, d1, m2, d2):
  '''
  month_apart(m1, d1, m2, d2)
  m1-> month 1
  m2-> month 2
  d1-> days of month 1
  d2-> days of month 2

  Returns True if dates are at least a month apart, returns false if not.
  '''
  if abs(m1 - m2) > 1:
    return True
  elif m1 > m2 and (d1 - d2) >= 0:
    return True
  elif m1 < m2 and (d2 - d1) >= 0:
    return True
  else:
    return False
```
